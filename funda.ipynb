# Connect to WRDS and Obtain Dataset.

```python
import matplotlib.pyplot as plt
import wrds
db = wrds.Connection()
```

```python
db.list_libraries()
```

```python
db.list_tables(library="comp")
```

```python
db.describe_table(library="comp", table="funda")
```

```python
db.describe_table(library="comp", table="funda")['name'].to_list()
```

No "sic" and "gsubind" variables but offering from the webpage request?

```python
comp = db.raw_sql("""
                    select gvkey, tic, cusip, conm, datadate, fyear, sich, 
                    at, aqc, ppegt, ppent, emp, capx, capxv, sale, cogs, xad, xrd, xsga,
                    pstkl, txditc,
                    pstkrv, seq, pstk, mkvalt
                    from comp.funda
                    where indfmt='INDL'
                    and datafmt='STD'
                    and popsrc='D'
                    and consol='C'
                    and curcd='USD'
                    and datadate >= '01/01/1990'
                    and fic='USA'
                    and final='Y'
                    """)
```

```python
comp.to_csv("funda.csv", index=False)
```
